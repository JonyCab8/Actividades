```python
#ACTIVIDAD 3
#HORA: N1
#ALUMNO: JONATHAN RAMON CABRERA
#MATRICULA: 2020547

#Librerias
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt

#Conjunto de datos definido
np.random.seed(0)
horas_estudio = np.random.rand(100, 1) * 10
calificaciones = 2 * horas_estudio + np.random.randn(100, 1) * 2

#Seleccion y entrenamiento de modelo
train_data, test_data, train_labels, test_labels = train_test_split(horas_estudio, calificaciones, test_size=0.2, random_state=42)
modelo = LinearRegression().fit(train_data, train_labels)

predicciones = modelo.predict(test_data)
mse = mean_squared_error(test_labels, predicciones)
print("error cuadratico medio:", mse)

#Modelo de regresion lineal
plt.scatter(test_data, test_labels, color='green')
plt.plot(test_data, predicciones, color='blue')
plt.title('Modelo de Regresi√≥n Lineal')
plt.ylabel('Calificaciones')
plt.xlabel('Horas de Estudio')
plt.show()
```
